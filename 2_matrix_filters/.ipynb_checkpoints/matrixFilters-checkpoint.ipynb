```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import sparse
import sys


def showTheChannel(channel = 'all', path = 'test_img.jpg', rtrn = 1): # Отображает выбранный канал RGB
    img = plt.imread(path)
    copy = np.copy(img)
    
    if channel == 'r':
        copy[:, :, [1,2]] = 0
    elif channel == 'g':
        copy[:, :, [0,2]] = 0
    elif channel == 'b':
        copy[:, :, [0,1]] = 0
    elif channel == 'grey':
        grey = copy[:, :, [0]] / 3 + \
               copy[:, :, [1]] / 3 + \
               copy[:, :, [2]] / 3
        copy[:, :, [0, 1, 2]] = grey
    elif 'all':
        copy[:, :, :]
        
    if rtrn == 0:
        return copy 
    elif rtrn == 1:
        return plt.imshow(copy)
    
    
def blurFilter(kernelSize = 5, numCycles = 3, step = 1, path = 'test_img.jpg'): # Фильтр свёртки размытия
    img = plt.imread(path)
    copy = np.copy(img)
    fig = plt.figure(figsize = (15, 10))
    h = int(kernelSize - 1)
    
    for i in range(numCycles):
        for x in range(0, len(copy[:, 0, 0]) - \
                       h, step):
            for y in range(0, len(copy[0, :, 0]) - \
                           h, step):
                copy[x : x+h, y : y+h, 0] = \
                np.round(np.mean(copy[x : x+h, \
                                      y : y+h, 0]))
                copy[x : x+h, y : y+h, 1] = \
                np.round(np.mean(copy[x : x+h, \
                                      y : y+h, 1]))
                copy[x : x+h, y : y+h, 2] = \
                np.round(np.mean(copy[x : x+h, \
                                      y : y+h, 2]))
    plt.subplot(121)
    plt.imshow(img)
    plt.subplot(122)
    plt.imshow(copy)
    return plt.imshow(img), plt.imshow(copy)
    
    
def contrastFilter(N = 20, path = 'test_img.jpg'): # Изменение контрастности изображения
    img = plt.imread(path)
    copy = np.array(img, dtype = int)
    fig = plt.figure(figsize = (15, 10))
    middle = np.round(np.mean(copy[:, :, 0] + copy[:, :, 1] + copy[:, :, 2]))
    
    for x in range(0, len(copy[:, 0, 0])):
        for y in range(0, len(copy[0, :, 0])):
            for z in range(len(copy[x, y, :])):
                
                if copy[x, y, z] > middle:
                    copy[x, y, z] = copy[x, y, z] - N
                elif copy[x, y, z] < middle:
                    copy[x, y, z] = copy[x, y, z] + N
                    
                if copy[x, y, z] > 255:
                    copy[x, y, z] = 255
                elif copy[x, y, z] < 0:
                    copy[x, y, z] = 0
    plt.subplot(121)
    plt.imshow(img)
    plt.subplot(122)
    plt.imshow(copy)
    return  plt.imshow(img), plt.imshow(copy)


def borderlineFilter(path, mask = 'Sobel', rtrn = 0): # Фильтры Собеля и Превитта
    copy = np.copy(path)
    fig = plt.figure(figsize = (15, 10))
    
    if mask == 'Sobel':
        v = 2
        step = 3
    elif mask == 'Previtt':
        v = 1
        step = 3

    arrx = np.array([[-1,-v,-1],
                      [0, 0, 0],
                      [1, v, 1]])
    arry = np.array([[-1, 0, 1],
                     [-v, 0, v],
                     [-1, 0, 1]])  
    
    for x in range(0, len(copy[:, 0, 0]) - 2, step):
        for y in range(0, len(copy[0, :, 0]) - 2, step):
            px = np.sum(copy[x : x+3, y : y+3, 0] * arrx)
            py = np.sum(copy[x : x+3, y : y+3, 0] * arry)
            
            copy[x : x+3, y : y+3, :] = abs(px) + abs(py)
    plt.subplot(121)
    if rtrn == 0:
        return copy 
    elif rtrn == 1:
        return plt.imshow(copy)

    
def sparseBorderline(borderlinePath):
    copy = np.copy(borderlinePath)
    middle = np.round(np.mean(copy[:, :, 0] + copy[:, :, 1] + copy[:, :, 2]))
    for x in range(0, len(copy[:, 0, 0])):
        for y in range(0, len(copy[0, :, 0])):
            for z in range(len(copy[x, y, :])):
                if copy[x, y, z] >= middle:
                    copy[x, y, z] = 255
                elif copy[x, y, z] < middle:
                    copy[x, y, z] = 0
    return plt.imshow(copy), copy


def histogramClustering(path, clusters = 2): # Работает на основе гистаграммы
    copy = np.array(path, dtype = int)
    zrs = np.zeros(path.shape, dtype = int)
    
    for x in range(0, len(copy[:, 0, 0])):
        for y in range(0, len(copy[0, :, 0])):
            middle = np.round(np.mean(copy[x, y, 0] + copy[x, y, 1] + copy[x, y, 2]))
            zrs[x, y, :] = middle
            
    minimum = np.min(zrs)
    maximum = np.max(zrs)
    step = np.round((maximum - minimum) / clusters)
    start = 0
    end = 0
    
    for i in range(clusters):
        start = minimum + i * step
        end = start + step
        
        r = np.random.randint(255)
        g = np.random.randint(255)
        b = np.random.randint(255)
        
        for x in range(0, len(zrs[:, 0, 0])):
            for y in range(0, len(zrs[0, :, 0])):
                if zrs[x, y, 0] > start and zrs[x, y, 0] <= end:
                    zrs[x, y, [0]] = r
                    zrs[x, y, [1]] = g
                    zrs[x, y, [2]] = b
    return plt.imshow(zrs)


def thresholdBinarization(path, Threshold = 0, rtrn = 1): # Работает на основе гистаграммы
    copy = np.array(path, dtype = int)
    bins = np.arange(0, 255, 1)
    zrs = np.zeros(bins.shape)
    maxone = posone = maxtwo = postwo = 0

    for x in range(0, len(copy[:, 0, 0])):
        for y in range(0, len(copy[0, :, 0])):
            for i in range(0, len(bins)):
                if copy[x, y, 0] == bins[i]:
                    zrs[i] += 1

    for i in range(len(zrs)):
        if zrs[i] >= maxone:
            maxone = zrs[i]
            posone = i
    for i in range(int(posone / 2)):
        if zrs[i] >= maxtwo:
            maxtwo = zrs[i]
            postwo = i    

    binThreshold = posone - postwo + Threshold

    for x in range(0, len(copy[:, 0, 0])):
        for y in range(0, len(copy[0, :, 0])):
            for z in range(len(copy[x, y, :])):
                if copy[x, y, z] >= binThreshold:
                    copy[x, y, z] = 0
                elif copy[x, y, z] < binThreshold:
                    copy[x, y, z] = 255
    if rtrn == 0:
        return plt.hist(zrs, bins)
    elif rtrn == 1:
        return plt.imshow(copy)
    
def sparseRepr(a, rtrn = 1):
    x = list()
    for i in range(len(a)):
        aS = sparse.coo_matrix((a[i]), dtype=np.int32)
        x.append((i, aS))
    if rtrn == 1:
        return sys.getsizeof(a), sys.getsizeof(x)
    elif rtrn == 0:
        return x
```

7.3. Сегментация изображения методом порогового ограничения (число кластеров = 10). Выбор границ кластера осуществляется разбиением гистограммы на равные доли
